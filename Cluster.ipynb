```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
digits=load_digits()
d=pd.DataFrame(digits.data)
d ["Target"]=digits.target
print(d.head())
```

```python
kmeans=KMeans(n_clusters=10,random_state=42)
kmeans.fit(digits.data)
```

```python
d["cluster"]=kmeans.labels_
print(d.head())
```

```python
pca=PCA(n_components=2)
pca_result=pca.fit_transform(digits.data)
```

```python
plt.scatter(pca_result[:,0],pca_result[:,1],c=d["cluster"],cmap="viridis")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.title("PCA Results")
plt.show()
```

```python
cl=kmeans.cluster_centers_.reshape(10,8,8)
fig,axes=plt.subplots(2,5,figsize=(8,3))
for i,ax in enumerate(axes.flat):
    ax.imshow(cl[i],cmap="gray")
    ax.axis("off")
plt.show()
```
